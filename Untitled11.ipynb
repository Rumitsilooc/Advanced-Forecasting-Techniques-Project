```python
# ==============================
# 1. Install required libraries
# ==============================
!pip install gdown pandas numpy statsmodels scikit-learn matplotlib

# ==============================
# 2. Imports
# ==============================
import gdown
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ==============================
# 3. Download dataset from Google Drive
# ==============================
# Your Google Drive file link:
# https://drive.google.com/file/d/1HXidHvHwvYUFL6CAxXZwlF4Y_vltZks3/view?usp=drive_link

file_id = "1HXidHvHwvYUFL6CAxXZwlF4Y_vltZks3"
url = f"https://drive.google.com/uc?id={file_id}"

output = "Walmart_Data_Analysis_and_Forcasting.csv"
gdown.download(url, output, quiet=False)

# ==============================
# 4. Load and prepare data
# ==============================
df = pd.read_csv(output)

# Convert Date to datetime and sort
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df = df.sort_values(['Store', 'Date'])

print("Dataset shape:", df.shape)
print("\nColumns and dtypes:")
print(df.dtypes)

print("\nMissing values per column:")
print(df.isna().sum())

print("\nDate range:", df['Date'].min(), "→", df['Date'].max())
print("Number of stores:", df['Store'].nunique())

# ==============================
# 5. Select one store and build SARIMAX
# ==============================
store_id = 1   # you can change this to other store numbers
store_df = df[df['Store'] == store_id].copy()

# Set Date as index
store_df = store_df.set_index('Date')

# Target and exogenous variables
y = store_df['Weekly_Sales']
exog = store_df[['Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']]

# ==============================
# 6. Train/Test split
# ==============================
test_size = 20  # last 20 weeks as test set
y_train, y_test = y.iloc[:-test_size], y.iloc[-test_size:]
exog_train, exog_test = exog.iloc[:-test_size], exog.iloc[-test_size:]

print("\nTrain period:", y_train.index.min(), "→", y_train.index.max())
print("Test period :", y_test.index.min(), "→", y_test.index.max())

# ==============================
# 7. Fit SARIMAX model
# ==============================
order = (1, 1, 1)
seasonal_order = (1, 1, 1, 52)  # yearly seasonality for weekly data

sarimax_model = SARIMAX(
    y_train,
    exog=exog_train,
    order=order,
    seasonal_order=seasonal_order,
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarimax_result = sarimax_model.fit(disp=False)
print("\n=== SARIMAX Summary ===")
print(sarimax_result.summary())

# ==============================
# 8. Forecast on test set
# ==============================
forecast = sarimax_result.get_forecast(steps=test_size, exog=exog_test)

y_pred = forecast.predicted_mean
conf_int = forecast.conf_int()

# Align indices
y_pred.index = y_test.index
conf_int.index = y_test.index

# ==============================
# 9. Evaluation metrics
# ==============================
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(f"\n=== SARIMAX performance for Store {store_id} ===")
print("MAE :", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("MAPE:", round(mape, 2), "%")

# ==============================
# 10. Plot actual vs forecast
# ==============================
plt.figure(figsize=(12, 5))
plt.plot(y_train[-40:], label='Train (last 40 weeks)')
plt.plot(y_test, label='Actual (test)', marker='o')
plt.plot(y_pred, label='SARIMAX forecast', marker='o')

plt.fill_between(
    conf_int.index,
    conf_int.iloc[:, 0],
    conf_int.iloc[:, 1],
    alpha=0.2,
    label='95% CI'
)

plt.title(f"Store {store_id} – Weekly Sales: Actual vs SARIMAX Forecast")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Make sure Date is datetime and sorted
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df = df.sort_values(['Store', 'Date'])

stores = df['Store'].unique()
print("Number of stores:", len(stores))

results = []

test_size = 20  # last 20 weeks as test set
order = (1, 1, 1)
seasonal_order = (1, 1, 1, 52)

for store_id in stores:
    store_df = df[df['Store'] == store_id].copy()
    store_df = store_df.set_index('Date')

    y = store_df['Weekly_Sales']
    exog = store_df[['Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']]

    # Skip if not enough data
    if len(y) <= test_size + 10:
        print(f"Skipping Store {store_id}: not enough data")
        continue

    y_train, y_test = y.iloc[:-test_size], y.iloc[-test_size:]
    exog_train, exog_test = exog.iloc[:-test_size], exog.iloc[-test_size:]

    try:
        model = SARIMAX(
            y_train,
            exog=exog_train,
            order=order,
            seasonal_order=seasonal_order,
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        result = model.fit(disp=False)

        forecast = result.get_forecast(steps=test_size, exog=exog_test)
        y_pred = forecast.predicted_mean

        # Metrics
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

        results.append({
            "Store": store_id,
            "MAE": mae,
            "RMSE": rmse,
            "MAPE": mape
        })

        print(f"Store {store_id}: MAPE={mape:.2f}%  RMSE={rmse:.1f}")

    except Exception as e:
        print(f"Store {store_id}: error -> {e}")

# Convert to DataFrame
results_df = pd.DataFrame(results)
print("\nPer-store SARIMAX performance:")
print(results_df)

print("\nAverage performance across stores:")
print(results_df[['MAE', 'RMSE', 'MAPE']].mean())
```

to predict next week’s Weekly_Sales.

Tested on the last 20 weeks (same test horizon as SARIMAX for Store 1).

Uses a 2-layer LSTM (hidden size 64) trained with MSE loss and Adam optimize

```python
# ==============================
# 1. Install and import libraries
# ==============================
!pip install gdown pandas numpy torch scikit-learn matplotlib --quiet

import gdown
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ==============================
# 2. Download and load the data
# ==============================
file_id = "1HXidHvHwvYUFL6CAxXZwlF4Y_vltZks3"
url = f"https://drive.google.com/uc?id={file_id}"
output = "Walmart_Data_Analysis_and_Forcasting.csv"
gdown.download(url, output, quiet=False)

df = pd.read_csv(output)

# ==============================
# 3. Basic preprocessing
# ==============================
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df = df.sort_values(['Store', 'Date'])

print("Dataset shape:", df.shape)
print(df.head())

# ==============================
# 4. Focus on one store (Store 1)
# ==============================
store_id = 1
store_df = df[df['Store'] == store_id].copy()
store_df = store_df.sort_values('Date').reset_index(drop=True)

print(f"\nUsing Store {store_id}, length:", len(store_df))
print(store_df.head())

# ==============================
# 5. Select features and target
# ==============================
feature_cols = ['Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']
target_col = 'Weekly_Sales'

data = store_df[feature_cols].values  # features + target (first column)
dates = store_df['Date'].values

# ==============================
# 6. Scale features (including target separately)
# ==============================
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_all = data[:, 1:]           # exogenous features
y_all = data[:, 0:1]          # Weekly_Sales as column vector

X_scaled = scaler_X.fit_transform(X_all)
y_scaled = scaler_y.fit_transform(y_all)

# Concatenate back: [scaled_target, scaled_features]
data_scaled = np.concatenate([y_scaled, X_scaled], axis=1)

# ==============================
# 7. Build sequences for LSTM
# ==============================
seq_len = 12        # weeks of history
test_size = 20      # last 20 weeks as test

N = len(data_scaled)
cutoff_idx = N - test_size    # index where test starts (in original point indices)

X_seq_train, y_seq_train = [], []
X_seq_test, y_seq_test = [], []
dates_test = []

for i in range(N - seq_len):
    x_seq = data_scaled[i:i+seq_len, :]       # 12 weeks of all features
    y_target_idx = i + seq_len                # predict next week
    y_target = data_scaled[y_target_idx, 0]   # scaled Weekly_Sales

    if y_target_idx < cutoff_idx:
        X_seq_train.append(x_seq)
        y_seq_train.append(y_target)
    else:
        X_seq_test.append(x_seq)
        y_seq_test.append(y_target)
        dates_test.append(dates[y_target_idx])

X_seq_train = np.array(X_seq_train)
y_seq_train = np.array(y_seq_train)
X_seq_test = np.array(X_seq_test)
y_seq_test = np.array(y_seq_test)
dates_test = np.array(dates_test)

print("\nTrain sequences:", X_seq_train.shape, "Test sequences:", X_seq_test.shape)

# ==============================
# 8. PyTorch Dataset & DataLoader
# ==============================
class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = TimeSeriesDataset(X_seq_train, y_seq_train)
test_dataset = TimeSeriesDataset(X_seq_test, y_seq_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# ==============================
# 9. Define LSTM model
# ==============================
class LSTMForecaster(nn.Module):
    def __init__(self, n_features, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=n_features,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        # x: [batch, seq_len, n_features]
        out, _ = self.lstm(x)
        # take last time step
        out = out[:, -1, :]      # [batch, hidden_size]
        out = self.fc(out)       # [batch, 1]
        return out.squeeze(1)    # [batch]

n_features = data_scaled.shape[1]   # 1 target + 5 exogenous = 6

model = LSTMForecaster(n_features=n_features, hidden_size=64, num_layers=2).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# ==============================
# 10. Train the LSTM
# ==============================
epochs = 40

for epoch in range(1, epochs+1):
    model.train()
    train_losses = []
    for X_batch, y_batch in train_loader:
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())

    if epoch % 5 == 0 or epoch == 1:
        print(f"Epoch {epoch}/{epochs}, Train Loss: {np.mean(train_losses):.4f}")

# ==============================
# 11. Evaluate on test set
# ==============================
model.eval()
y_pred_scaled = []

with torch.no_grad():
    for X_batch, _ in test_loader:
        X_batch = X_batch.to(device)
        outputs = model(X_batch)
        y_pred_scaled.extend(outputs.cpu().numpy())

y_pred_scaled = np.array(y_pred_scaled)
y_true_scaled = y_seq_test

# Inverse scale to original Weekly_Sales
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
y_true = scaler_y.inverse_transform(y_true_scaled.reshape(-1, 1)).flatten()

# Compute metrics
mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print(f"\n=== LSTM performance for Store {store_id} (next-week forecast) ===")
print("MAE :", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("MAPE:", round(mape, 2), "%")

# ==============================
# 12. Plot actual vs predicted for test period
# ==============================
plt.figure(figsize=(12,5))
plt.plot(dates_test, y_true, label="Actual", marker='o')
plt.plot(dates_test, y_pred, label="LSTM Forecast", marker='o')
plt.title(f"Store {store_id} – LSTM Next-Week Forecast (Test Period)")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#
# ==============================
# 2. Dataset helper + LSTM model
# ==============================
class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]


class LSTMForecaster(nn.Module):
    def __init__(self, n_features, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=n_features,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)      # [batch, seq_len, hidden]
        out = out[:, -1, :]        # last time step
        out = self.fc(out)         # [batch, 1]
        return out.squeeze(1)

# ==============================
# 3. Loop over all stores
# ==============================
stores = df['Store'].unique()
print("Number of stores:", len(stores))

seq_len = 12        # history length (weeks)
test_size = 20      # last 20 weeks for testing
epochs = 25         # keep moderate; increase if you have time
batch_size = 32
results = []

for store_id in stores:
    print(f"\n=== Store {store_id} ===")
    store_df = df[df['Store'] == store_id].copy()
    store_df = store_df.sort_values('Date').reset_index(drop=True)

    if len(store_df) <= seq_len + test_size + 5:
        print("Not enough data, skipping.")
        continue

    feature_cols = ['Weekly_Sales', 'Holiday_Flag', 'Temperature',
                    'Fuel_Price', 'CPI', 'Unemployment']
    target_col = 'Weekly_Sales'
    data = store_df[feature_cols].values

    # scale per store
    scaler_X = StandardScaler()
    scaler_y = StandardScaler()

    X_all = data[:, 1:]      # exogenous
    y_all = data[:, 0:1]     # Weekly_Sales

    X_scaled = scaler_X.fit_transform(X_all)
    y_scaled = scaler_y.fit_transform(y_all)
    data_scaled = np.concatenate([y_scaled, X_scaled], axis=1)

    N = len(data_scaled)
    cutoff_idx = N - test_size

    X_seq_train, y_seq_train = [], []
    X_seq_test, y_seq_test = [], []

    for i in range(N - seq_len):
        x_seq = data_scaled[i:i+seq_len, :]
        y_idx = i + seq_len
        y_val = data_scaled[y_idx, 0]   # scaled Weekly_Sales

        if y_idx < cutoff_idx:
            X_seq_train.append(x_seq)
            y_seq_train.append(y_val)
        else:
            X_seq_test.append(x_seq)
            y_seq_test.append(y_val)

    X_seq_train = np.array(X_seq_train)
    y_seq_train = np.array(y_seq_train)
    X_seq_test = np.array(X_seq_test)
    y_seq_test = np.array(y_seq_test)

    print("Train seq:", X_seq_train.shape, "Test seq:", X_seq_test.shape)

    train_dataset = TimeSeriesDataset(X_seq_train, y_seq_train)
    test_dataset = TimeSeriesDataset(X_seq_test, y_seq_test)

    train_loader = DataLoader(train_dataset, batch_size=batch_size,
                              shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size,
                             shuffle=False)

    n_features = data_scaled.shape[1]
    model = LSTMForecaster(n_features=n_features,
                           hidden_size=64,
                           num_layers=2).to(device)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    # ----- train -----
    for epoch in range(1, epochs+1):
        model.train()
        losses = []
        for Xb, yb in train_loader:
            Xb = Xb.to(device)
            yb = yb.to(device)

            optimizer.zero_grad()
            out = model(Xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            losses.append(loss.item())
        if epoch % 10 == 0 or epoch == 1:
            print(f"Epoch {epoch}/{epochs}, loss={np.mean(losses):.4f}")

    # ----- evaluate -----
    model.eval()
    y_pred_scaled = []
    with torch.no_grad():
        for Xb, _ in test_loader:
            Xb = Xb.to(device)
            out = model(Xb)
            y_pred_scaled.extend(out.cpu().numpy())

    y_pred_scaled = np.array(y_pred_scaled)
    y_true_scaled = y_seq_test

    y_pred = scaler_y.inverse_transform(
        y_pred_scaled.reshape(-1, 1)
    ).flatten()
    y_true = scaler_y.inverse_transform(
        y_true_scaled.reshape(-1, 1)
    ).flatten()

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    print(f"Store {store_id} LSTM -> MAE={mae:.1f}, RMSE={rmse:.1f}, MAPE={mape:.2f}%")

    results.append({
        "Store": store_id,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    })

# ==============================
# 4. Summarize results
# ==============================
results_df = pd.DataFrame(results)
print("\nPer-store LSTM performance:")
print(results_df)

print("\nAverage LSTM performance across stores:")
print(results_df[['MAE', 'RMSE', 'MAPE']].mean())
```

For individual stores, such as Store 1, the LSTM model was able to slightly outperform SARIMAX, suggesting that deep learning can capture additional non-linear patterns when sufficient data is available. ## GRU MOdel

```python
# ============================================
# GLOBAL GRU MODELS FOR WALMART DEMAND
# - GRU-12 (12-week history)
# - GRU-26 (26-week history)
# ============================================

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ============================================
# 1. Load and prepare data
# ============================================

# Adjust this path if needed
file_path = "Walmart_Data_Analysis_and_Forcasting.csv"
df = pd.read_csv(file_path)

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df = df.sort_values(['Store', 'Date']).reset_index(drop=True)

print("Data shape:", df.shape)
print(df.head())

feature_cols = ['Weekly_Sales', 'Holiday_Flag', 'Temperature',
                'Fuel_Price', 'CPI', 'Unemployment']
target_col = 'Weekly_Sales'

# ============================================
# 2. Dataset class
# ============================================

class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        # X: [n_samples, seq_len, n_features]
        # y: [n_samples]
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# ============================================
# 3. GRU model
# ============================================

class GRUForecaster(nn.Module):
    def __init__(self, n_features, hidden_size=64, num_layers=2):
        super().__init__()
        self.gru = nn.GRU(
            input_size=n_features,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        # x: [batch, seq_len, n_features]
        out, _ = self.gru(x)           # out: [batch, seq_len, hidden]
        out = out[:, -1, :]            # last time step: [batch, hidden]
        out = self.fc(out)             # [batch, 1]
        return out.squeeze(1)          # [batch]

# ============================================
# 4. Helper: build global sequences for all stores
# ============================================

def build_global_sequences(df, seq_len=12, test_size=20):
    """
    Build global train/test sequences for all stores.
    Last `test_size` weeks of each store are used as test.
    """
    stores = df['Store'].unique()
    X_train_raw, y_train_raw, store_train = [], [], []
    X_test_raw, y_test_raw, store_test = [], [], []

    for store_id in stores:
        sdf = df[df['Store'] == store_id].sort_values('Date')
        data = sdf[feature_cols].values  # [N, n_features]

        if len(data) <= seq_len + test_size + 5:
            # Not enough data for this store
            continue

        N = len(data)
        cutoff_idx = N - test_size  # y indices >= cutoff_idx belong to test

        for i in range(N - seq_len):
            x_seq = data[i:i+seq_len, :]      # window
            y_idx = i + seq_len              # index of prediction point
            y_val = data[y_idx, 0]           # Weekly_Sales (first column)

            if y_idx < cutoff_idx:
                X_train_raw.append(x_seq)
                y_train_raw.append(y_val)
                store_train.append(store_id)
            else:
                X_test_raw.append(x_seq)
                y_test_raw.append(y_val)
                store_test.append(store_id)

    X_train_raw = np.array(X_train_raw)   # [n_train, seq_len, n_features]
    y_train_raw = np.array(y_train_raw)   # [n_train]
    X_test_raw  = np.array(X_test_raw)    # [n_test, seq_len, n_features]
    y_test_raw  = np.array(y_test_raw)    # [n_test]
    store_train = np.array(store_train)
    store_test  = np.array(store_test)

    return X_train_raw, y_train_raw, store_train, X_test_raw, y_test_raw, store_test

# ============================================
# 5. Helper: scale data (fit on train, apply to both)
# ============================================

def scale_data(X_train_raw, y_train_raw, X_test_raw, y_test_raw):
    n_train, seq_len, n_features = X_train_raw.shape
    n_test = X_test_raw.shape[0]

    # Flatten temporal dimension for X to fit scaler
    X_train_flat = X_train_raw.reshape(-1, n_features)
    X_test_flat = X_test_raw.reshape(-1, n_features)

    scaler_X = StandardScaler()
    scaler_y = StandardScaler()

    X_train_scaled_flat = scaler_X.fit_transform(X_train_flat)
    X_test_scaled_flat  = scaler_X.transform(X_test_flat)

    y_train_scaled = scaler_y.fit_transform(y_train_raw.reshape(-1, 1)).flatten()
    y_test_scaled  = scaler_y.transform(y_test_raw.reshape(-1, 1)).flatten()

    # Reshape back to [n_samples, seq_len, n_features]
    X_train_scaled = X_train_scaled_flat.reshape(n_train, seq_len, n_features)
    X_test_scaled  = X_test_scaled_flat.reshape(n_test, seq_len, n_features)

    return X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled, scaler_y

# ============================================
# 6. Helper: train and evaluate one GRU
# ============================================

def train_and_evaluate_gru(seq_len, epochs=50, batch_size=64, test_size=20):
    print(f"\n===============================")
    print(f"GRU with seq_len={seq_len}, epochs={epochs}")
    print(f"===============================")

    # Build sequences
    X_train_raw, y_train_raw, store_train, X_test_raw, y_test_raw, store_test = \
        build_global_sequences(df, seq_len=seq_len, test_size=test_size)

    print("Train samples:", X_train_raw.shape[0], "| Test samples:", X_test_raw.shape[0])

    # Scale
    X_train, y_train, X_test, y_test, scaler_y = \
        scale_data(X_train_raw, y_train_raw, X_test_raw, y_test_raw)

    # DataLoaders
    train_dataset = TimeSeriesDataset(X_train, y_train)
    test_dataset  = TimeSeriesDataset(X_test, y_test)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader  = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    # Model
    n_features = X_train.shape[2]
    model = GRUForecaster(n_features=n_features, hidden_size=64, num_layers=2).to(device)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    # Training loop
    for epoch in range(1, epochs+1):
        model.train()
        losses = []
        for Xb, yb in train_loader:
            Xb = Xb.to(device)
            yb = yb.to(device)

            optimizer.zero_grad()
            out = model(Xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            losses.append(loss.item())

        if epoch == 1 or epoch % 10 == 0:
            print(f"Epoch {epoch}/{epochs} - Train Loss: {np.mean(losses):.4f}")

    # Evaluation
    model.eval()
    y_pred_scaled = []
    with torch.no_grad():
        for Xb, _ in test_loader:
            Xb = Xb.to(device)
            out = model(Xb)
            y_pred_scaled.extend(out.cpu().numpy())

    y_pred_scaled = np.array(y_pred_scaled)
    y_true_scaled = y_test

    # Inverse-transform target
    y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
    y_true = scaler_y.inverse_transform(y_true_scaled.reshape(-1, 1)).flatten()

    # Metrics (overall across all stores)
    mae  = mean_absolute_error(y_true, y_pred)
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    print(f"\nGRU (seq_len={seq_len}) performance (ALL STORES, last {test_size} weeks):")
    print("MAE :", round(mae, 2))
    print("RMSE:", round(rmse, 2))
    print("MAPE:", round(mape, 2), "%")

    # Optional: quick plot for first 40 test points
    plt.figure(figsize=(12,5))
    plt.plot(y_true[:40], label="Actual")
    plt.plot(y_pred[:40], label="GRU Forecast")
    plt.title(f"GRU seq_len={seq_len} – Sample of Test Forecasts")
    plt.xlabel("Sample index (across stores)")
    plt.ylabel("Weekly Sales")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return mae, rmse, mape

# ============================================
# 7. Run GRU-12 and GRU-26
# ============================================

mae_12, rmse_12, mape_12 = train_and_evaluate_gru(seq_len=12, epochs=50, batch_size=64, test_size=20)
mae_26, rmse_26, mape_26 = train_and_evaluate_gru(seq_len=26, epochs=50, batch_size=64, test_size=20)

print("\nSummary:")
print(f"GRU-12  -> MAE={mae_12:.1f},  RMSE={rmse_12:.1f},  MAPE={mape_12:.2f}%")
print(f"GRU-26  -> MAE={mae_26:.1f},  RMSE={rmse_26:.1f},  MAPE={mape_26:.2f}%")
```
